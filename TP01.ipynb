```python
Blue=np.array([255,255,255,255,0,0,0,0,0,0,0,0,255,255,255,255,255,255,255,255,0,0,0,0,255,255,255,255])
Green=np.array([0,0,0,0,255,255,255,255,0,0,0,0,255,255,255,255,255,255,255,255,255,255,255,255,0,0,0,0])
Red=np.array([0,0,0,0,0,0,0,0,255,255,255,255,255,255,255,255,0,0,0,0,255,255,255,255,255,255,255,255])
Blue=np.tile(Blue,(28,1))
Green=np.tile(Green,(28,1))
Red=np.tile(Red,(28,1))
Img=cv.merge([Red,Green,Blue])
Img=np.uint8(Img)
Img=cv.cvtColor(Img,cv.COLOR_BGR2RGB)
plt.imshow(Img)
```

L'image est codée sur 8 bits 
le poids theorique  de l'image est :  28*28 oct

```python
cv.imwrite('Colors.bmp',Img)
```
